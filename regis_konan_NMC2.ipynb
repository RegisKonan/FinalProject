Consider the normalized predator-prey (Lotka-Volterra) model given by :

\begin{align}\label{equ1}
\begin{cases}
    \frac{dB}{dt} =& B-BR \\    
    \frac{dR}{dt} =& -R+BR
\end{cases}
\end{align}
For $B_0=2, R_0=1$

## Define the Package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import warnings
import pylab as p
warnings.filterwarnings("ignore")
```

## Define the function predator-prey (Latka-volterra)

```python
def f(t,y):
    B,R = y
    dBdt = B-B*R
    dRdt = -R + B*R
    dydt = np.array([dBdt,dRdt])
    return dydt
```

## Euler fonction 

```python
def Euler(f, y0):
    n = len(t)
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(n - 1):
        h = t[i+1] - t[i]
        y[i+1] = y[i] +  h*f (t[i],y[i])
    return y
```

## Runge kutta order 4 fonction

```python
def rungekutta4(f, y0, t):
    n = len(t)
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(n - 1):
        h = t[i+1] - t[i]
        k1 = f(t[i],y[i])
        k2 = f(t[i] + h / 2.,y[i] + k1 * h / 2.)
        k3 = f(t[i] + h / 2.,y[i] + k2 * h / 2.)
        k4 = f(t[i] + h, y[i] + h*k3)
        y[i+1] = y[i] + (h / 6.) * (k1 + 2*k2 + 2*k3 + k4)
    return y
```

## Initial condition and time interval

```python
y0 =[2,1]
t = np.linspace(0, 20, 100)
sol1= Euler(f, y0)
sol2= rungekutta4(f, y0, t)
```

## Plot Euler and Runge Kutta

```python
fig=plt.figure(1, figsize=(6, 6))
plt.plot(t,sol1[:,0],'b',label='B')
plt.plot(t,sol1[:,1],'r',label='R')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title("Forward Euler")
plt.legend()
plt.show()
plt.plot(t,sol2[:,0],'b',label='B')
plt.plot(t,sol2[:,1],'r',label='R')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.title("Runge kutta")
plt.show()
```

### Plot the solution trajectories for Euler and Runge Kutta

```python
z1=sol1[:,0]
z2=sol1[:,1]
z3=sol2[:,0]
z4=sol2[:,1]
```

```python
plt.plot(z1,z2,'g',label=' Forward Euler')
plt.xlabel('Prey')
plt.ylabel('Predator')
plt.title("Forward Euler")
plt.show()
plt.plot(z3,z4,'b')
plt.xlabel('Prey')
plt.ylabel('Predator')
plt.title("Runge kutta")
plt.show()
```
