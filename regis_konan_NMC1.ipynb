```python
def Lorentz(t,G, sigma, r,b): 
    return np.array([D[0]*(G[1]-G[0]),D[1]*G[0]-G[1]-G[0]*G[2] ,G[0]*G[1]-D[2]*G[2]])
```

# DEFINE RUNGE KUTTA

```python
def rungekutta4(f, Y0, t, args=()):
    n = len(t)
    Y = np.zeros((n, len(Y0)))
    Y[0] = Y0
    for i in range(n - 1):
        h = t[i+1] - t[i]
        k1 = f(t[i],Y[i], *args)
        k2 = f(t[i] + h / 2.,Y[i] + k1 * h / 2., *args)
        k3 = f(t[i] + h / 2.,Y[i] + k2 * h / 2., *args)
        k4 = f(t[i] + h, Y[i] + h*k3, *args)
        Y[i+1] = Y[i] + (h / 6.) * (k1 + 2*k2 + 2*k3 + k4)
    return Y
```

```python
Y0 = np.array([8,0,30])
t = np.linspace(-20, 20, 1000)
sol = rungekutta4(Lorentz, Y0, t, args=(D))
```

# PLOT THE LORENTZ FUNCTION USING RUNGE KUTTA IN 3D

```python
fig=plt.figure(1, figsize=(5, 5))
ax=fig.gca(projection='3d')
ax.plot(sol[:,0],sol[:,1], sol[:,2],'g',label='Lorentz 3D Solution')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
```
