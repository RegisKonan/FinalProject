Plot the figure below

- The x range is from 0, to 100, with a step size of 1
- The function for the blue curve is exp(-0.01 * x)
- The function for the red curve is 0.1 + exp(-0.1*x)
- The black dashed line is just a constant with a value of 0.4


Set the font size to 20, color values are "tab:blue", "tab:red", and "black", if there are other values that you need to set, choose any appropriate values.



training.png

```python
############## Add your code here ##############
'''
General Steps:
1. Define the domain x.
2. Define the function values for each curve.
3. Make any changes you think are needed. 
'''
x=np.arange(0,100,1)
y_1=np.exp(-0.01*x)
y_2=0.1+np.exp(-0.1*x)
y_3=np.full(x.shape,0.4)
plt.plot(x,y_1,"b")
plt.plot(x,y_2,"r")
plt.plot(x,y_3,linestyle='dotted',color='black')
plt.xlabel("Training iteration",fontsize=20)
plt.ylabel("Training error",fontsize=20)
plt.title("Training Curve",fontsize=20)
plt.legend(["Model1","Model2","baseline"],fontsize=20)
plt.xticks(np.arange(0, 120,20),fontsize=20)
plt.yticks(np.arange(0.2, 1.2, 0.2),fontsize=20)

################################################
plt.show()
```

- Check the marker values in the documentation link: https://matplotlib.org/stable/api/markers_api.html

```python
############## Helper function to generate some radnom data from a gaussiain ##############
def generte_data(mean_x1, mean_x2, std, size):
  x1 = mean_x1 + std * np.random.randn(size)
  x2 = mean_x2 + std * np.random.randn(size)
  return x1, x2
###########################################################################################

# generate data for a scatter plot
x1, y1 = generte_data(0.2, 0.5, 0.1, 1000)
x2, y2 = generte_data(0.5, -0.5, 0.1, 1000)
x3, y3 = generte_data(0, 0, 0.1, 1000)

############## Add your code here ##############
'''
General Steps:
1. Generate the scatter plot for the data provided above.
2. Add the title, x and y labels.
3. Modify the legends, x and y ticks to match the plot above.
'''
plt.scatter(x1,y1,color='tab:blue',marker="o")
plt.scatter(x2,y2,color='tab:green',marker="x")
plt.scatter(x3,y3,color='tab:red',marker="^")


plt.title('Dummy Classification Problem',fontsize=20)
plt.xlabel('X',fontsize=20)
plt.ylabel('Y',fontsize=20)
plt.legend(['Class1', 'Class2', 'Class3'],fontsize=20)
plt.xticks(np.arange(-0.4, 0.8,0.2),fontsize=20)
plt.yticks(np.arange(-0.75, 1, 0.25),fontsize=20)
################################################
plt.show()
```
